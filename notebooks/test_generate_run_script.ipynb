```python
import os
import os.path as osp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

models = ['transe', 'distmult', 'complex', 'rotate']
datasets = ['cn15k', 'nl27k', 'ppi5k']
confi_score_functions = ['logi', 'rect']
fc = 'l1'
b = True


lrs = [0.01, 0.005, 0.001]
hidden_dims = [128, 256, 512]

count = {}
count[0] = 0
count[1] = 0
count[2] = 0

cuda_device_map = {0: 2, 1: 3, 2: 0}

for model in models:
    for dataset in datasets:
        for confi in confi_score_functions:
            for lr in lrs:
                for hidden_dim in hidden_dims:
                    path = f"/home/mou/Projects/UKGE-FL/results/unc_{dataset}_{model}_confi_{confi}_fc_{fc}_bias_{b}_dim_{hidden_dim}/lr_{lr}_wd_0.0_topk_True"
                    file_path  = osp.join(path, 'best_model_ndcg_lin.pth')
                    if not osp.exists(file_path):
                    #     print(f"Not working {dataset}, {model} {confi} {hidden_dim}: {lr} ")
                    # else:
                    #     print(f"####### Working {dataset}, {model} {confi} {hidden_dim}: {lr} ")
                        device = datasets.index(dataset)
                        count[device] += 1
                        print(f"CUDA_VISIBLE_DEVICES={cuda_device_map[device]} nohup python -u src/ukge/train_unc_baseline.py --model {model} --dataset {dataset} --confidence_score_function {confi} --fc_layers {fc} --bias --hidden_dim {hidden_dim} --lr {lr} --topk &> logs/unc_{model}_{dataset}_confi_{confi}_fc_{fc}_bias_{b}_lr_{lr}_dim_{hidden_dim}.log & ")

print(count)
                    
```
